```python
# Permutando os principio ativos compostos

from itertools import permutations

P_Ativo_combinacao=[]
P_Ativo_posicao = []

for index, val in enumerate(df.PRINCIPIO_ATIVO):
  if val.find('+')>=0:
    composicao = val.split('+')
    for y in range(len(composicao)):
      composicao[y] = composicao[y].strip()
    x = len(composicao)
    if x < 9:
      temp = permutations(composicao, x)
      for i in list(temp):
        for j in range(len(i)):
          if j == 0:
            k=i[j]
          else:
            k = k+' + '+i[j]
        P_Ativo_combinacao.append(k)
        P_Ativo_posicao.append(index)
    else:
      pass

#print(P_Ativo_combinado)
```
